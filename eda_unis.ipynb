```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
unis = pd.read_excel('./data/list_unis.xlsx')
```

```python
unis.head()
```

```python
unis.drop(labels = 'Unnamed: 0', axis = 1, inplace = True)
```

```python
unis['name_uni'] = unis['name_uni'].apply(lambda x: x.title())
unis.head()
```

```python
unis["plz"] = unis["plz"].convert_dtypes(int)
```

```python
# Drop Gebühren 
unis.drop("s_gebuehren", axis = 1, inplace= True)
unis.drop("s_beitrag", axis = 1, inplace= True)
```

```python
unis.head()
```

```python
unis["sd_nm"].unique()
```

```python
percentage_column_list = ['sd_mg',	'sd_nm',	'sd_rws',	'sd_sk',	'sd_in',	'sd_so',	'ab_ba',	'ab_ma',	'ab_la',	'ab_dms',	'ab_so']
```

```python
def to_num_or_null(x):
    try:
        return pd.to_numeric(x)
    except:
        return np.nan
```

```python
unis['total_students'] = unis['total_students'].apply(lambda x : to_num_or_null(str(x).replace(".", "")))
unis['total_students'] = unis['total_students'].convert_dtypes(int)
```

```python
unis['total_students_loc'] = unis['total_students_loc'].apply(lambda x : to_num_or_null(str(x).replace(".", "")))
unis['total_students_loc'] = unis['total_students_loc'].convert_dtypes(int)
```

```python
unis['total_students'].unique()
```

```python
for element in percentage_column_list:
    unis[element] = unis[element].apply(lambda x : to_num_or_null(str(x).replace(" %", "").replace(",","."))/100)
```

```python
unis.drop("founding_year", axis = 1, inplace= True)
```

```python
unis.head()
```

```python
unis.rename(columns = {'sd_mg' : 'med_ges_percentage'}, inplace = True)
unis.rename(columns = {'sd_nm' : 'nat_math_percentage'}, inplace = True)
unis.rename(columns = {'sd_rws' : 'law_econ_social_percentage'}, inplace = True)
unis.rename(columns = {'sd_sk' : 'lang_cult_percentage'}, inplace = True)
unis.rename(columns = {'sd_in' : 'engineer_inf_percentage'}, inplace = True)
unis.rename(columns = {'sd_so' : 'other_percentage'}, inplace = True)
```

```python
unis.rename(columns = {'ab_ba' : 'bachelor_percentage'}, inplace = True)
unis.rename(columns = {'ab_ma' : 'master_percentage'}, inplace = True)
unis.rename(columns = {'ab_la' : 'lehramt_percentage'}, inplace = True)
unis.rename(columns = {'ab_dms' : 'dipl_magis_staatsex_percentage'}, inplace = True)
unis.rename(columns = {'ab_so' : 'other_percentage_grad'}, inplace = True)
```

```python
unis.head()
```

```python
unis.columns
```

```python
students_per_sd = ['med_ges_percentage', 'nat_math_percentage',
       'law_econ_social_percentage', 'lang_cult_percentage',
       'engineer_inf_percentage', 'other_percentage', 'bachelor_percentage',
       'master_percentage', 'lehramt_percentage',
       'dipl_magis_staatsex_percentage', 'other_percentage_grad']
    
```

```python

unis['total_med_ges'] = unis.apply(lambda x: x.med_ges_percentage * x.total_students_loc if x.total_students_loc == int else x.med_ges_percentage * x.total_students, axis=1)
unis['total_nat_math'] = unis.apply(lambda x: x.nat_math_percentage * x.total_students_loc if x.total_students_loc == int else x.nat_math_percentage * x.total_students, axis=1)
unis['total_law_econ_social'] = unis.apply(lambda x: x.law_econ_social_percentage * x.total_students_loc if x.total_students_loc == int else x.law_econ_social_percentage * x.total_students, axis=1)
unis['total_lang_cult'] = unis.apply(lambda x: x.lang_cult_percentage * x.total_students_loc if x.total_students_loc == int else x.lang_cult_percentage * x.total_students, axis=1)
unis['total_engineer_info'] = unis.apply(lambda x: x.engineer_inf_percentage * x.total_students_loc if x.total_students_loc == int else x.engineer_inf_percentage * x.total_students, axis=1)
unis['total_other'] = unis.apply(lambda x: x.other_percentage * x.total_students_loc if x.total_students_loc == int else x.other_percentage * x.total_students, axis=1)
unis['total_bachelor'] = unis.apply(lambda x: x.bachelor_percentage * x.total_students_loc if x.total_students_loc == int else x.bachelor_percentage * x.total_students, axis=1)
unis['total_master'] = unis.apply(lambda x: x.master_percentage * x.total_students_loc if x.total_students_loc == int else x.master_percentage * x.total_students, axis=1)
unis['total_lehramt'] = unis.apply(lambda x: x.lehramt_percentage * x.total_students_loc if x.total_students_loc == int else x.lehramt_percentage * x.total_students, axis=1)
unis['total_dipl_magis_staatsex'] = unis.apply(lambda x: x.dipl_magis_staatsex_percentage * x.total_students_loc if x.total_students_loc == int else x.dipl_magis_staatsex_percentage * x.total_students, axis=1)
unis['total_other_grad'] = unis.apply(lambda x: x.other_percentage_grad * x.total_students_loc if x.total_students_loc == int else x.other_percentage_grad * x.total_students, axis=1)
```

```python
unis["total_med_ges"].fillna(0, inplace=True)
```

```python
unis["total_med_ges"] = unis["total_med_ges"].astype(int)
```

```python
totals = ['total_nat_math','total_law_econ_social','total_lang_cult','total_engineer_info','total_other','total_bachelor','total_master','total_lehramt','total_dipl_magis_staatsex','total_other_grad']
def converty(x):
    unis[x].fillna(0, inplace=True)
    unis[x] = unis[x].astype(int)
for c in totals:
    converty(c)  
unis.head()
```

```python
unis.to_csv("./data/unis_cleaned.csv")
unis.to_excel("./data/unis_cleaned.xlsx")
```
