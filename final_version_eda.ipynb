```python
def definer_mg(x):
    x = x.lower() 
    if  x.find('medi') > -1 or x.find('medi') > -1 or x.find('gesund') > -1 or x.find('pharma') > -1 or x.find('spo') > -1 or x.find('psy') > -1  or x.find('lebe') > -1 or x.find('human') > -1 or x.find('leben') > -1:
        return 1
    else:
        return 0
def definer_mn(x):
    x = x.lower() 
    if x.find('math') > -1 or x.find('phy') > -1 or x.find('tech') > -1 or x.find('chem') > -1 or x.find('bio') > -1 or x.find('mint') > -1 or x.find('nat') > -1 or x.find('geo') > -1:
        return 1
    else:
        return 0
def definer_rws(x): 
    x = x.lower() 
    if x.find('wirt') > -1 or x.find('sozial') > -1 or x.find('recht') > -1  or x.find('jur') > -1  or x.find('betr') > -1 or x.find('gese') > -1 or x.find('pol') > -1  or x.find('ver') > -1 or x.find('geist') > -1 or x.find('volksw') > -1 or x.find('bwl') > -1:
        return 1
    else:
        return 0
def definer_sk(x):
    x = x.lower()  
    if x.find('engl') > -1 or x.find('eng.') > -1 or x.find('eng-') > -1 or x.find('phil') > -1 or x.find('phil') > -1 or x.find('franz') > -1 or x.find('amer') > -1 or x.find('angl') > -1  or x.find('jap') > -1 or x.find('asia') > -1 or x.find('lingu') > -1 or x.find('kultur') > -1 or x.find('literat') > -1:
        return 1
    else:
        return 0
def definer_ii(x):
    x = x.lower() 
    if x.find('ingenieur') > -1 or x.find('info') > -1 or x.find('tech') > -1 or x.find('masch') > -1 or x.find('elektro') > -1 or x.find('archit') > -1 or x.find('bau') > -1 or x.find('luft') > -1 or x.find('raum') > -1 :
        return 1
    else:
        return 0
```
